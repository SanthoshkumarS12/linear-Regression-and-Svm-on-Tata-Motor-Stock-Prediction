# PCA

```python
DS =pd.read_csv(r"C:\Users\santh\OneDrive\Desktop\studies\machine learning\mlproject\TATAMOTORS.csv")
```

```python
DS
```

```python
X = DS.iloc[: , 3:11].values  
Y = DS.iloc[: , 11:12].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)  
```

```python
SC = StandardScaler()   
X_train = SC.fit_transform(X_train) 
print(X_train)
X_test = SC.transform(X_test)  
print(X_test)
```

```python
pca =PCA(n_components=2)
pct=pca.fit_transform(X_train,X_test)
principal_df=pd.DataFrame(pct,columns=['PC1','PC2'])
final_df=pd.concat([principal_df,DS[['Turnover']]],axis=1)
```

```python
print('before pca\n----------------------------------------------------')
print(DS[:5])
print('after pca\n----------------------------------------------------')
print(final_df[:5])
```
